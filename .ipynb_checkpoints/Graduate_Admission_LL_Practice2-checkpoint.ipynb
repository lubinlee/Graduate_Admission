```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import statsmodels.api as sm
import sys
import os
import seaborn as sns

%matplotlib inline
```

```python
df = pd.read_csv("Admission_Predict_Ver1.1.csv")
```

```python
df=df.rename(columns = {'Chance of Admit ':'Chance of Admit'})
```

```python
df.rename(columns={'Serial No.':'Srno',
                   'GRE Score':'GRE',
                   'TOEFL Score':'TOEFL',
                   'University Rating':'UnivRating',
                   'Chance of Admit':'Chance'},inplace=True)
```

```python
df.head()
```

```python
df.drop('Srno', axis=1, inplace=True)
df.head()
```

```python
#correlations_data = admt.corr()['Chance'].sort_values(ascending=False)
cor=df.corr()['Chance']
# Print the correlations
print(cor)
```

GRE and Research on Chance of Admission

```python
X = df[['GRE','Research']]
y = df['Chance']
```

```python
from sklearn.cross_validation import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.4, random_state=101)
```

```python
from sklearn.linear_model import LinearRegression
lm = LinearRegression()
lm.fit(X_train,y_train)
```

```python
predictions = lm.predict(X_test)
```

```python
plt.scatter(y_test,predictions)
```

GRE, TOEFL and Research on Chance of Admission

```python
X = df[['GRE','Research', 'TOEFL']]
y = df['Chance']
```

```python
from sklearn.cross_validation import train_test_split
from sklearn.linear_model import LinearRegression

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.4, random_state=101)

lm = LinearRegression()
lm.fit(X_train,y_train)

predictions = lm.predict(X_test)
plt.scatter(y_test,predictions)
```

GRE, TOEFL, CGPA and Research on Chance of Admission

```python
X = df[['GRE','Research', 'TOEFL', 'CGPA']]
y = df['Chance']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.4, random_state=101)

lm = LinearRegression()
lm.fit(X_train,y_train)

predictions = lm.predict(X_test)
plt.scatter(y_test,predictions)
```

Another way of training and testing and getting linear regression results in number

```python
from sklearn.metrics import mean_squared_error
```

```python
models = [['Linear Regression :', LinearRegression()]]

print("Results...")

for name, model in models:
    model = model
    model.fit(X_train, y_train)
    predictions = model.predict(X_test)
    print(name, (np.sqrt(mean_squared_error(y_test, predictions))))
```
